# Data Exercise TSV file

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\friesco\workstation\fr-p\studies\ASTRO716\data_excercise\psdr1_new.tsv"

try:
    df = pd.read_csv(file_path, delimiter='\t', comment='#', header=None)
    first_row = df.iloc[0]
    if '\t' in first_row.str.cat(sep='\t'):
        df = pd.read_csv(file_path, delimiter='\t', comment='#', header=0)
    else:
        cols = ['ID', '_r', '_RAJ2000', '_DEJ2000', 'RAJ2000', 'DEJ2000', 'objID', 'f_objID', 'Qual', 'Ns', 'Nd', 'gmag', 'e_gmag', 'gFlags', 'rmag', 'e_rmag', 'rFlags', 'imag', 'e_imag', 'iFlags', 'zmag', 'e_zmag', 'zFlags']
        df.columns = cols
except FileNotFoundError:
    print(f"Error: '{file_path}' not found.")
    exit()
except pd.errors.ParserError:
    print("Error: Parsing failed.")
    exit()
except Exception as e:
    print("Error:", e)

print("Data Summary:")
print(df.info())
print("\nFirst 5 rows:\n", df.head())
print("\nDescriptive Statistics:\n", df.describe())

if 'objID' in df.columns:
    num_objs = df['objID'].nunique()
    print(f"\nNumber of unique objects: {num_objs}")

if 'objInfoFlag' in df.columns:
    print("\nObject Type Distribution:")
    print(df['objInfoFlag'].value_counts())

filters = ['g', 'r', 'i', 'z', 'y']
for f in filters:
    if f'{f}mag' in df.columns:
        plt.figure()
        plt.hist(df[f'{f}mag'].dropna(), bins=50)
        plt.title(f'{f}-band Magnitudes')
        plt.xlabel(f'{f}mag')
        plt.ylabel('Frequency')
        plt.show()

if 'rmag' in df.columns and 'imag' in df.columns:
    plt.figure()
    plt.scatter(df['rmag'], df['imag'], s=5, alpha=0.5)
    plt.title('r vs. i Magnitudes')
    plt.xlabel('r mag')
    plt.ylabel('i mag')
    plt.show()
```
